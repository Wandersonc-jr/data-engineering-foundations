```python
# Importing libraries

import numpy as np
import pandas as pd
```

Reading .csv file

```python
df = pd.read_csv("data/Clientes.csv", delimiter =',' )
df
```

Looking for "Mato Grosso" in the 'state' column

```python
mt_state = df.loc[df['state'] == 'Mato Grosso']
mt_state
```

Looking for a list of states

```python
states = ['Mato Grosso','Bahia', 'São Paulo']

df_states = df.loc[df['state'].isin(states)]
df_states
```

Handling NaNs and nulls in the 'street' and 'number' columns

```python
df['street'] = df['street'].fillna('Não registrado')
df['number'] = pd.to_numeric(df['number'], errors='coerce')
df['number'] = df['number'].fillna(0).astype(int)
df
```

Getting info about 'first_name' and 'additionals'

```python
df[['first_name','additionals']].info()
```

Handling NaNs in the 'additionals'

```python
df['additionals'] = df['additionals'].fillna('Sem complemento')
df
```
