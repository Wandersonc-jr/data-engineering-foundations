```python
# Importing libraries
import pandas as pd
```

Reading .xlsx file

```python
df = pd.read_excel("data/Produto.xlsx")
df
```

looking for 'Calça' on column 'Name'

```python
df_pants = df.loc[df['Name'] == 'Calça']
df_pants
```

Looking for a list of items in the 'Name' column

```python
itens = ['Calça', 'TV','Sapato vermelho', 'Fogão']
df_itens = df.loc[df['Name'].isin(itens)]
df_itens
```

Showing prices that are greater than 1000

```python
df_price_greater_than_1000 = df.loc[df['Price'] > 1000]
df_price_greater_than_1000
```

Showing prices that are less than 1000

```python
df_price_less_than_1000 = df.loc[df['Price'] < 1000]
df_price_less_than_1000
```

Showing prices between 200 and 1000


```python
df_200_to_1000 = df.loc[(df['Price'] > 200) & (df['Price'] <= 1000)]
df_200_to_1000
```

Showing only 'Name', 'Price' from items whose price is between 200 and 1000

```python
df_200_to_1000[['Name','Price']]
```
